```python
#5
rem = re.compile(r'[^\d\.]')
us_census["Income"] = [rem.sub("",a) for a in us_census["Income"]]
us_census["Hispanic"] = [rem.sub("",x) for x in us_census["Hispanic"]]
us_census["White"] = [rem.sub("",x) for x in us_census["White"]]
us_census["Black"] = [rem.sub("",x) for x in us_census["Black"]]
us_census["Native"] = [rem.sub("",x) for x in us_census["Native"]]
us_census["Asian"] = [rem.sub("",x) for x in us_census["Asian"]]
us_census["Income"]=us_census["Income"].astype(float)
us_census.head()
```

```python
#6
us_census[["Men","Women"]]=us_census.GenderPop.str.split("_",expand=True)
us_census.head()
```

```python
#7
rem = re.compile(r'[^\d]')
us_census["Women"] = [rem.sub("",x) for x in us_census["Women"]]
us_census["Men"] = [rem.sub("",x) for x in us_census["Men"]]
us_census.head()
```

```python
#8
plt.scatter(us_census["Women"] , us_census["Income"] ) 
plt.show() 
```

```python
#9
us_census["Men"]=us_census["Men"].astype(int)
rem=re.compile(r"^\s*$")
us_census["Women"].replace(rem,np.nan,regex=True,inplace=True)
filler=us_census["TotalPop"]-us_census["Men"]
us_census["Women"].fillna(filler,inplace=True)
us_census["Women"]=us_census["Women"].astype(int)
us_census["Women"]
```

```python
#14
us_census["Hispanic"] = us_census["Hispanic"].astype(float)
us_census["White"] = us_census["White"].astype(float)
us_census["Black"] = us_census["Black"].astype(float)
us_census["Native"] = us_census["Native"].astype(float)
us_census["Asian"] = us_census["Asian"].astype(float)
rem=re.compile(r"[/%]")
us_census["Pacific"] =us_census["Pacific"].str.replace(rem,"").astype(float)
filler=us_census["Pacific"].mean()
us_census["Pacific"].fillna(filler, inplace=True)
us_census=us_census.drop_duplicates()
```
